```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

def generar_funcion_periodica(T, num_terminos):
    # Coeficiente a_0 (trmino constante)
    a_0 = np.random.uniform(-1, 1)

    # Coeficientes a_n y b_n para n = 1, 2, ..., num_terminos
    a_n = np.random.uniform(-1, 1, num_terminos)
    b_n = np.random.uniform(-1, 1, num_terminos)

    # Dominio de la funcin peridica
    x = np.linspace(0, T, 1000)

    # Inicializar la funcin peridica con el trmino constante a_0
    f_x = a_0 / 2

    # Agregar la contribucin de los trminos coseno y seno
    for n in range(1, num_terminos + 1):
        f_x += a_n[n - 1] * np.cos(2 * np.pi * n * x / T) + b_n[n - 1] * np.sin(2 * np.pi * n * x / T)

    return x, f_x

# Parmetros para la funcin peridica
periodo = 10   # Tamao del perodo
num_terminos = 10  # Nmero de trminos en la serie de Fourier

# Generar la funcin peridica
x, f_x = generar_funcion_periodica(periodo, num_terminos)

# Graficar la funcin peridica
plt.figure(figsize=(10, 5))
plt.plot(x, f_x)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Funcin Peridica Aleatoria')
plt.grid(True)
plt.show()

#Exportacin de la grafica en TXT
files.download('funcion_periodica.txt')
```
