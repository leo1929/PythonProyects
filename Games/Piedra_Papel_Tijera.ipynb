# JUEGO PIEDRA PAPEL O TIJERA

El juego de piedra, papel o tijera es un juego de manos y decisiones en el que dos jugadores eligen uno de los tres elementos: piedra, papel o tijera, con el objetivo de vencer al oponente. La mecánica básica del juego es la siguiente:

1. La piedra vence a las tijeras: si un jugador elige piedra y el otro elige tijera, la piedra rompe las tijeras y el jugador que eligió piedra gana.
2. El papel vence a la piedra: si un jugador elige papel y el otro elige piedra, el papel envuelve la piedra y el jugador que eligió papel gana.
Las tijeras vencen al papel: si un jugador elige tijera y el otro elige papel, las tijeras cortan el papel y el jugador que eligió tijera gana.
3. En caso de que ambos jugadores elijan el mismo elemento, se considera un empate y se suele jugar otra ronda. El juego se basa en la idea de que cada elemento tiene una relación de dominancia sobre otro y una relación de debilidad con respecto al tercero.

Rock, paper and scissors es un juego muy popular y se utiliza para tomar decisiones de manera aleatoria o divertida, ya sea entre amigos, en competencias o incluso en situaciones más formales, como desempates en competiciones deportivas o en el ámbito de la toma de decisiones en algunos procesos.

```python
#Lipieza del la ventana
import os
os.system('clear')

#Juego piedra papel o tijera
import random
options = ('piedra', 'papel', 'tijera')

#Contador de del puntaje acumulado
computer_wins = 0
leo_wins = 0
count = 0

#Inicialización del número de rondas
rounds = 1

while True:
    print('*' * 45)
    print(' -*-*-*-*-*-*-*-*- ROUND', rounds,'-*-*-*-*-*-*-*-*-')
    print('*' * 45)
    print('' * 20)

    #Mensaje para el contador del puntaje
    print('    Computer_wins:', computer_wins, '||', 'Leo_wins:', leo_wins)
    print('' * 20)

    #Opción variable:
    print('-*-*-*-*-*-*-*-*-[  SELECT  ]-*-*-*-*-*-*-*-*-')
    print('' * 20)
    user_option = input('Piedra, Papel o Tijera: ')
    user_option = user_option.lower()
    rounds += 1

    #Condición para evitar un valor invalido
    if not user_option in options:
      print('' * 20)
      print(' ********* Opcion no es valida *********')
      print('' * 20)
      #Este comando me omite todo lo que este debajo de el y vuelve al iniciar un round
      continue

    #Selección usauario vs PC
    computer_option = random.choice(options)
    print('' * 20)
    print('1. Leo option      =>', user_option)
    print('2. Computer option =>', computer_option)
    print('' * 20)

    #                MENU
    if user_option == computer_option:
        print('        /<>/<>/<>/ Empate /<>/<>/<>/')
        print('' * 20)
    elif user_option == 'piedra':
        if computer_option == 'tijera':
            print('********* Piedra gana a Tijera *********')
            print('' * 20)
            print('     /<>/<>/<>/ Leo gano /<>/<>/<>/')
            print('' * 20)
            leo_wins += 1
        else:
            print('********* Papel gana a piedra *********')
            print('' * 20)
            print('   /<>/<>/<>/ Computer gano /<>/<>/<>/')
            print('' * 20)
            computer_wins += 1
    elif user_option == 'papel':
        if computer_option == 'piedra':
            print('********* Papel gana a Piedra *********')
            print('' * 20)
            print('     /<>/<>/<>/ Leo gano /<>/<>/<>/')
            print('' * 20)
            leo_wins += 1
        else:
            print('********* Tijera gana a Papel *********')
            print('' * 20)
            print(' /<>/<>/<>/ Computer gano /<>/<>/<>/')
            print('' * 20)
            computer_wins += 1
    elif user_option == 'tijera':
        if computer_option == 'papel':
            print('********* Tijera gana a Papel *********')
            print('' * 20)
            print('     /<>/<>/<>/ Leo gano /<>/<>/<>/ ')
            print('' * 20)
            leo_wins += 1
        else:
            print('********* Piedra gana a Tijera *********')
            print('' * 20)
            print('   /<>/<>/<>/ Computer gano /<>/<>/<>/')
            print('' * 20)
            computer_wins += 1

    if computer_wins == 3:
      print('********* El ganador es la computadora *********')
      print('' * 20)
      print('    Computer_wins:', computer_wins, '||', 'Leo_wins:', leo_wins)
      print('' * 20)
      count = 1
      break

    if leo_wins == 3:
      print('********* El ganador es Leo *********')
      print('' * 20)
      print('    Computer_wins:', computer_wins, '||', 'Leo_wins:', leo_wins)
      print('' * 20)
      count = 1
      break
```
